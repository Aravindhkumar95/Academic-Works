```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


data = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/00471/Data_for_UCI_named.csv")
```

```python
data
```

```python
X = data.iloc[:,:-2]
Y = data.iloc[:,-2]
```

# PRINCIPLE COMPONENT ANALYSIS

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=4)
XPCA = pca.fit_transform(X)
```

```python
XPCA = pd.DataFrame(XPCA)
XPCA
```

```python
from sklearn.model_selection import train_test_split
XPCA_train, XPCA_test, Y_train, Y_test = train_test_split(XPCA,Y,test_size=0.25, random_state=0)
```

# FIRST REGRESSION APPROACH IN PCA
#             RIDGE REGRESSION

```python
from sklearn.preprocessing import scale
from sklearn.linear_model import Ridge, RidgeCV
from sklearn.metrics import mean_squared_error
import time
```

```python
alphas = 10**np.linspace(10,-2,100)*0.5
alphas
```

```python
ridge = Ridge(normalize=True)
coefs = []
for a in alphas:
    ridge.set_params(alpha=a)
    ridge.fit(XPCA, Y)
    coefs.append(ridge.coef_)
np.shape(coefs)
```

```python
ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('coef')
```

```python
ridge_1 = Ridge(alpha=10, normalize=True)
ridge_1.fit(XPCA_train, Y_train) 
pred_1 = ridge_1.predict(XPCA_test) 
print(pd.Series(ridge_1.coef_, index=XPCA.columns))
print(mean_squared_error(Y_test, pred_1))
```

```python
ridge_2 = Ridge(alpha=10**10, normalize=True)
ridge_2.fit(XPCA_train, Y_train)
pred_2 = ridge_2.predict(XPCA_test)
print(pd.Series(ridge_2.coef_, index=XPCA.columns)) 
print(mean_squared_error(Y_test, pred_2))
```

```python
ridge_3 = Ridge(alpha=0.1, normalize=True)

time_start = time.clock()
ridge_3.fit(XPCA_train, Y_train)
time_elapsed_PCA_RG1train = (time.clock() - time_start)

time_start = time.clock()
pred_3 = ridge_3.predict(XPCA_test) 
time_elapsed_PCA_RG1test = (time.clock() - time_start)

print(pd.Series(ridge_3.coef_, index=XPCA.columns)) 
print(mean_squared_error(Y_test, pred_3)) 
```

```python
ridgecv = RidgeCV(alphas=alphas, scoring=None, normalize=True)
ridgecv.fit(XPCA_train, Y_train)
ridgecv.alpha_
```

```python
ridge_4 = Ridge(alpha=ridgecv.alpha_, normalize=True)
ridge_4.fit(XPCA_train, Y_train)
mean_squared_error(Y_test, ridge_4.predict(XPCA_test))
```

```python
mse_pca_rr = mean_squared_error(Y_test, pred_3)
```

```python
ridge_4.fit(XPCA, Y)
pd.Series(ridge_4.coef_, index=XPCA.columns)
```

```python
Mean_Squeared_Error_PCA_Ridge = mean_squared_error(Y_test, pred_3)
```

# SECOND REGRESSION APPROACH IN PCA
#             LASSO REGRESSION

```python
from sklearn.linear_model import Lasso, LassoCV

lasso = Lasso(max_iter=10000, normalize=True)
coefs = []
for a in alphas:
    lasso.set_params(alpha=a)
    lasso.fit(scale(XPCA_train), Y_train)
    coefs.append(lasso.coef_)

ax = plt.gca()
ax.plot(alphas, coefs)
ax.set_xscale('log')
plt.axis('tight')
plt.xlabel('alpha')
plt.ylabel('coef')
```

```python

lassocv = LassoCV(alphas=None, cv=10, max_iter=100000, normalize=True)
lassocv.fit(XPCA_train, Y_train)
lasso.set_params(alpha=lassocv.alpha_)


time_start = time.clock()
lasso.fit(XPCA_train, Y_train)
time_elapsed_PCA_RG2train = (time.clock() - time_start)

time_start = time.clock()
lass_pred1 = lasso.predict(XPCA_test)
time_elapsed_PCA_RG2test = (time.clock() - time_start)

mean_squared_error(Y_test, lass_pred1)
```

```python
pd.Series(lasso.coef_, index=XPCA.columns)
```

```python
Mean_Squeared_Error_PCA_Lasso = mean_squared_error(Y_test, lass_pred1)
```

```python
#Correlation with output variable

cor_target = abs(cor["stab"])

#Selecting highly correlated features

relevant_features = cor_target[cor_target>0.29]
relevant_features
```
