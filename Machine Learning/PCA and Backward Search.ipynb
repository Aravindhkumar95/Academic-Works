```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
mean1 = [-3, -2, -1, -2, 3, 5, 8, 7, 5, 2]
mean2 = [3, 1, 2, 0, -2, -3, -5, -6, -8, -9]


cov1=np.array([[3,9,21,27,57,0,0,0,0,0],[9,7,32,64,0,91,71,19,34,4],[21,32,11,6,31,49,41,43,0,2],[27,64,6,15,0,0,0,0,0,0],
              [57,0,31,0,19,0,0,0,0,0],[0,91,49,0,0,23,0,0,0,0],[0,71,41,0,0,0,28,0,0,0],[0,19,43,0,0,0,0,31,0,0],
              [0,34,0,0,0,0,0,0,35,0],[0,4,2,0,0,0,0,0,0,39]])


cov1
```

```python
x1 = np.random.multivariate_normal(mean1, cov1, 500)
x2 = np.random.multivariate_normal(mean2, cov1, 500)
X = pd.DataFrame(np.concatenate((x1,x2)))
X.shape
print(pd.DataFrame(X))
```

```python
Xc = np.zeros(500)
Xc = np.concatenate((Xc, np.ones(500)))
```

```python
mu = np.mean(X)
Xm = X - mu
Xm
```

```python
e_val, e_vec = np.linalg.eig(np.dot(Xm.T,Xm))
print("Eigen Values")
print(pd.DataFrame(e_val))
print("Eigen Vectors")
print(pd.DataFrame(e_vec))
```

```python
idx = e_val.argsort()[::-1]   
e_val = e_val[idx]
e_vec = e_vec[:,idx]
```

```python
print("Eigen Values")
print(pd.DataFrame(e_val))
print("Eigen Vectors")
print(pd.DataFrame(e_vec))
```

# Backward Search

## Reducing the dimensionality to 9

```python
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

Y9 = X[9]
lreg = LinearRegression()
rfe = RFE(lreg, 9)
rfe = rfe.fit(X, Y9)

print('\n\nFEATUERS SELECTED\n\n')
print(rfe.support_)

w = 0

for i in range(0,9):
    if rfe.support_[i] != 1:
        w = w + i
    
X9b = X.drop(w,axis=1)
X9b.columns = [0,1,2,3,4,5,6,7,8]
X9b
```

```python
lda = sklearn.discriminant_analysis.LinearDiscriminantAnalysis()
lda.fit(X9b,Xc)
prediction = lda.predict(X9b)
acc_9b=accuracy_score(prediction,Xc)
error_9b = sum(abs(prediction - Xc))
error_b = []
error_b.append(error_9b)
print("accuracy", acc_9b)
print("total error", error_9b)
```

# Comparison of Classificaiton result of PCA and BACKWARD SERACH methods

```python
Backward search works by eliminating the least important features. 
It continues recursively until the specified number of features is reached. 

PCA (Principle Component Analysis) is a dimensionality reduction technique that projects the data into a lower dimensional space.



Since backward search eliminates the least features column the classification error of this method is more when compare to
the classification error of PCA method.
```
